```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image in grayscale
img = cv2.imread('car.jpg', cv2.IMREAD_GRAYSCALE)

# Convert the image to float32 for Harris corner detection
img_float32 = np.float32(img)

# Define Harris corner detection parameters
block_size = 2  # Neighborhood size for corner detection
ksize = 3  # Aperture parameter for Sobel operator
k = 0.04  # Harris corner detection sensitivity parameter

# Apply Harris corner detection
dst = cv2.cornerHarris(img_float32, block_size, ksize, k)

# Normalize the Harris response for visualization
dst_norm = cv2.normalize(dst, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Threshold the normalized response to find the corners
threshold_value = 100  # Adjust this threshold to control the number of detected corners
corners_img = np.zeros_like(img)
corners_img[dst_norm > threshold_value] = 255

# Display the original image and the detected corners
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(corners_img, cmap='gray')
plt.title('Harris Corner Detection')
plt.axis('off')

plt.show()
```
